```python
# List comprehension
long_list = [i for i in range(30)]
print(long_list)

long_long_list = [(i, j) for i in range(3) for j in range(4)]
print(long_long_list)

long_list_list = [[i for i in range(3)] for _ in range(4)] # true representation of the matrix
print(long_list_list)

matrix = "\n".join([" ".join(["0" for x in range(4)]) for x in range(4)])
print(matrix)
```

# Dictionary and Set

```python
# Set (unordered, unique)
set_1 = {i ** 2 for i in range(10)}
print(set_1)
```

```python
print([i ** 2 for i in range(-10, 10)])
print({i ** 2 for i in range(-10, 10)})
```

```python
# Dictionary (mapping)
dict_1 = {(5-i): i ** 2 for i in range(10)}
print(dict_1)
print(dict_1.keys())
```

```python
# Dictionary update:
dict_2 = {'a': 10, 'b': 11}
dict_1.update(dict_2)
dict_1
```

```python
# Iterate through items:
for k, v in dict_1.items():
    print(k, v)
```

```python
import numpy as np
A = np.array([[1,2,3],[11,12,13],[21,22,23]])
x = np.array([[1],[2],[3]])
y = np.array([[1,2,3]])

print(A.T)                   # Transpose of A.
print(A*x)                   # Matrix multiplication of A and x.
print(np.invert(A))          # Inverse of A.
print(np.linalg.solve(A, x)) # Solve the linear equation system.


print("==========================================")
from numpy import matrix
from numpy import linalg
A = matrix( [[1,2,3],[11,12,13],[21,22,23]]) # Creates a matrix.
x = matrix( [[1],[2],[3]] )                  # Creates a matrix (like a column vector).
y = matrix( [[1,2,3]] )                      # Creates a matrix (like a row vector).
print(A.T)                                   # Transpose of A.
print(A*x)                                   # Matrix multiplication of A and x.
print(A.I)                                   # Inverse of A.
print(linalg.solve(A, x))                    # Solve the linear equation system.
```

## Numpy - Broadcasting
Numpy compares dimensions of operands, then infer s missing/mismatched
dimensions so the operation is still valid. Be careful with DIMENSIONS

```python
import numpy as np
import pprint as pp

op1 = np.array([i for i in range(9)]).reshape(3,3)
op2 = np.array([[1,2,3]])
op3 = np.array([1,2,3])

# the results are different
pp.pprint(op1 + op2)
pp.pprint(op1 + op2.T)

# the results are the same
pp.pprint(op1 + op3)
pp.pprint(op1 + op3.T)
```

## Vectorize is better option for computatioin

```python
import numpy as np

a = np.random.random(5000000)
b = np.random.random(5000000)
```

```python
# using loop
%%time
dot = 0.0
for i in range(len(a)):
    dot += a[i] * b[i]
print(dot)
```

```python
# using vectorization
%%time
print(np.array(a).dot(np.array(b)))
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create data
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(t, s)

# Format the plot
ax.set(xlabel='time (s)', ylabel='voltage (mV)', title='Sample plot')
ax.grid()

# save/show
fig.savefig('sample.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# create data
x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# plot the data
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis lable')
plt.ylabel('y axis lable')
plt.title('Sine and Consine Plot')
plt.legend(['Sine', 'Cosine'])
plt.show()
```
